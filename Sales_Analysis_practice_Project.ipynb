```python
# Sales Performance Analysis
#
# A simple statistical analysis of weekly sales data to understand:
# - Central tendency
# - Variability
# - Business insights
```

```python

# -------------------------------
# Data Section
# -------------------------------
# Sample weekly sales data (in units)
sales = [500, 700, 600, 800, 900, 750, 650]
ads_spend = [50, 70, 60, 80, 90, 75, 65]
products_sold = [
    "sunflower", "broccoli", "sunflower",
    "basil", "sunflower", "broccoli", "sunflower"
]
```

```python
# -------------------------------
# Import Libraries
# -------------------------------
import statistics

# -------------------------------
# Core Calculations
# -------------------------------
```

```python
# Central tendency
# Calculate average sales
mean_sales = statistics.mean(sales)
median_sales = statistics.median(sales)
# Spread / variability
sales_range = max(sales) - min(sales)
variance_sales = statistics.pvariance(sales)
std_dev_sales = statistics.pstdev(sales)
```

```python
# Most frequently sold product
most_sold_product = max(set(products_sold), key=products_sold.count)
```

```python
# Relationship between ads and sales
correlation = statistics.correlation(ads_spend, sales)
```

```python
# -------------------------------
# Output Section
# -------------------------------
```

```python
print("\n--- Sales Analysis Summary ---\n")

print(f"Average Sales        : {mean_sales:.2f}")
print(f"Median Sales         : {median_sales}")
print(f"Sales Range          : {sales_range}")
print(f"Variance (Population): {variance_sales:.2f}")
print(f"Standard Deviation   : {std_dev_sales:.2f}")
```

```python

print("\n--- Business Insights ---\n")

print(f"Top-Selling Product  : {most_sold_product}")
print(f"Ads vs Sales Correlation: {correlation:.2f}")
```

```python
print("Insight: There is a strong positive relationship between advertising spend and sales, suggesting that increasing marketing budget may drive higher revenue.")
```
